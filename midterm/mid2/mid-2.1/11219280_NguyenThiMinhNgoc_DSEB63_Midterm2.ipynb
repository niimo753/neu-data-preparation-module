```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Exercise 1

```python
df1 = pd.read_csv("data/traffic_by_month.csv")
df1
```

* Create a line chart for this data
* Remove the left, right, top spine
* Remove the ticks located at the bottom
* Color the xtick labels and the spine to grey
* Annotate some data points
* Add title and subtitle for this chart

```python
fig1, ax1 = plt.subplots()

# Create a line chart for this data
xdata = df1['month']
ydata = df1['traffic_volume']
ax1.plot(xdata, ydata, color='grey')

# Remove the left, right, top spine
ax1.spines[['left', 'right', 'top']].set_visible(False)

# Remove the ticks located at the bottom
ax1.tick_params(bottom=False)

# Color the xtick labels and the spine to grey
ax1.tick_params(axis='x', labelcolor='grey')
ax1.spines[['bottom']].set_color('grey')

# Annotate some data points
lowest3 = df1.sort_values(by='traffic_volume')[:3]
for x, y in zip(lowest3['month'], lowest3['traffic_volume']):
    ax1.annotate(text=int(y),
                 xy=(x,y),
                 xytext=(5, -5),
                 textcoords="offset points")
highest3 = df1.sort_values(by='traffic_volume', ascending=False)[:3]
for x, y in zip(highest3['month'], highest3['traffic_volume']):
    ax1.annotate(text=int(y),
                 xy=(x,y),
                 xytext=(1, 5),
                 textcoords="offset points",
                 ha='center')
ax1.plot(lowest3['month'], lowest3['traffic_volume'], '.k')
ax1.plot(highest3['month'], highest3['traffic_volume'], '.k')

# Add title and subtitle for this chart
ax1.text(s="Traffic Volume By Month",
         x=ax1.get_xlim()[0]-1.5,
         y=ax1.get_ylim()[1]+35,
         fontsize='xx-large',
         fontweight='bold')
ax1.text(s="Traffic Volume By Month",
         x=ax1.get_xlim()[0]-1.45,
         y=ax1.get_ylim()[1],
         fontsize='medium',
         fontweight='light')

# For more beautiful graph
ax1.set_xticks(ticks=range(1, 13), labels=range(1, 13))
ax1.set_xlabel("Month")
```

# Exercise 2

```python
df2 = pd.read_csv("data/sales.csv")
df2
```

```python
fig2, ax2 = plt.subplots(figsize=(9, 5), nrows=2, ncols=1, sharex=True)

# The first subplot - line chart
ax2[0].plot(df2['DateTime'], df2['NbOfSalesEmployees'], marker='o')
ax2[0].set(ylim=(60, 120))
ax2[0].spines[['bottom', 'top', 'right']].set_visible(False)
ax2[0].spines[['left']].set_color("grey")
ax2[0].tick_params(axis='both', bottom=False, color='grey', labelcolor='grey')
ax2[0].set_yticks([60, 80, 100, 120], labels=[60, 80, 100, 120])
ax2[0].set_ylabel('# of Sales\n Employees', color='grey', loc='top', fontsize='large')

# The second subplot - bar chart
ax2[1].bar(df2['DateTime'], df2['Revenue(Millions)'])
ax2[1].spines[['top', 'right']].set_visible(False)
ax2[1].tick_params(axis='both', bottom=False, color='grey', labelcolor='grey')
ax2[1].spines[['left', 'bottom']].set_color("grey")
ax2[1].set_yticks(np.arange(0, 1.5, 0.2))
ax2[1].set_yticklabels([f'${round(x,1)}' for x in ax2[1].get_yticks()])
ax2[1].set_ylabel('Revenue (Millions)', color='grey', loc='top', fontsize='large')
```
