```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime as dt
```

# Load Dataset

# Data Cleaning

```python
# rename some columns
exchange_rates = exchange_rates.rename(columns={r'Period\Unit:': 'Time', '[US dollar ]': 'US_dollar'})

# change the datatype of datetime column
exchange_rates['Time'] = pd.to_datetime(exchange_rates['Time'])

# change the datatype of datetime column and sort data in ascending order
exchange_rates = exchange_rates.sort_values(by='Time').reset_index(drop=True)
```

```python
# extract data (euros to US dollars currency exchange rate)
euro_to_dollar = exchange_rates[['Time', 'US_dollar']]

# number of each value in US_dollar columns
euro_to_dollar['US_dollar'].value_counts()
```

As we can see, there is 62 `-` values in `US_dollar` column, which represent unknown values. Hence, we will drop all the rows containing those unknown values.

```python
# drop rows that contain unknown value
euro_to_dollar = euro_to_dollar[euro_to_dollar['US_dollar']!='-']
```

This project will mainly focus on the EUR/USD exchange rate, so we will isolate the `Time` and the `US_dollar` columns, then assign them to a different variable named `euro_to_dollar`.

```python
# change the data type of US_dollar column to float
euro_to_dollar['US_dollar'] = euro_to_dollar['US_dollar'].astype('float')
```

```python
euro_to_dollar
```

# General Visualization

```python
fig1, ax1 = plt.subplots()

ax1.plot(euro_to_dollar['Time'], euro_to_dollar['US_dollar'], color='#B2B2B2')

ax1.text(s="EUR/USD Currency Exchange Rate",
        x=ax1.get_xlim()[0],
        y=ax1.get_ylim()[1]+0.07,
        fontsize='x-large',
        fontweight='bold')
ax1.text(s="EUR/USD daily exchange rate over the period from 1999 to 2023",
        x=ax1.get_xlim()[0],
        y=ax1.get_ylim()[1]+0.02,
        fontsize='medium',
        fontweight='light')
```

```python
fig1.savefig('euro-dollar-dailyer-all.png')
```

If we look at the line’s shape, we see many small wiggles – rather than seeing a smooth line. The wiggles, however, have meaning: they are the visual representation of the daily variation in the exchange rate. The rate goes up and down, up and down again, day to day. The rate only shows clear upward or downward trends in the longer run (months or years).

To show only the general long-term trends of the EUR/US currency exchange rate instead of showing the daily variation, we can use the rolling mean (moving average)

```python
# create a new column name rolling mean (using the moving windows of 30 days)
euro_to_dollar['rolling_mean'] = euro_to_dollar['US_dollar'].rolling(30).mean()
```

```python
fig2, ax2 = plt.subplots()

ax2.plot(euro_to_dollar['Time'], euro_to_dollar['rolling_mean'], color='#B2B2B2')
ax2.text(s="EUR/USD Currency Exchange Rate",
        x=ax1.get_xlim()[0],
        y=ax1.get_ylim()[1]+0.07,
        fontsize='x-large',
        fontweight='bold')
ax2.text(s="EUR/USD daily exchange rate over the period from 1999 to 2023",
        x=ax1.get_xlim()[0],
        y=ax1.get_ylim()[1]+0.02,
        fontsize='medium',
        fontweight='light')
```

```python
fig2.savefig('euro-dollar-rollmean-all.png')
```

# During The Coronavirus Pandemic

In this data analysis, we will mainly focus on the the way currency exchange rate between Euro and US dollar has changed during the coronavirus pandemic (2020), with the baseline years of 2016 - 2019. By using line plot, we can figure out the trend and some highlights in the EUR//USD exchange rate over this period.

```python
euro_to_dollar_16_19 = euro_to_dollar.loc[(euro_to_dollar['Time'].dt.year>=2016) & 
                                          (euro_to_dollar['Time'].dt.year<=2019)]
euro_to_dollar_20 = euro_to_dollar.loc[euro_to_dollar['Time'].dt.year==2020]
```

```python
fig3, ax3 = plt.subplots(3, figsize=(8, 8))

ax3[0].plot(euro_to_dollar['Time'], euro_to_dollar['rolling_mean'],
         label='1999-2023', color='grey', alpha=0.5, linewidth=0.5)
ax3[0].plot(euro_to_dollar_16_19['Time'], euro_to_dollar_16_19['rolling_mean'],
         label='2016 - 2019', color='grey', alpha=0.5)
ax3[0].plot(euro_to_dollar_20['Time'], euro_to_dollar_20['rolling_mean'],
         label='2020', color='black')
ax3[0].text(s="EUR/USD Currency Exchange Rate",
        x=ax3[0].get_xlim()[0],
        y=ax3[0].get_ylim()[1]+0.15,
        fontsize='x-large',
        fontweight='bold')
ax3[0].text(s="EUR/USD exchange rate over the period from 1999 to 2023",
        x=ax3[0].get_xlim()[0],
        y=ax3[0].get_ylim()[1]+0.05,
        fontsize='medium',
        fontweight='light')

ax3[0].axvspan(xmin=euro_to_dollar.loc[euro_to_dollar['Time'].dt.year==2016]['Time'].min(),
            xmax=euro_to_dollar.loc[euro_to_dollar['Time'].dt.year==2020]['Time'].max(),
            alpha=1, color='#E5E5E5')
ax3[0].axvspan(xmin=euro_to_dollar.loc[euro_to_dollar['Time'].dt.year==2020]['Time'].min(),
            xmax=euro_to_dollar.loc[euro_to_dollar['Time'].dt.year==2020]['Time'].max(),
            alpha=1, color='#B2B2B2')

ax3[0].text(s='2020', x=ax3[0].get_xlim()[0]+8200, y=1.4, color='black', alpha=1,
            bbox={'boxstyle': 'Round', 'facecolor':'white', 'edgecolor':'black', 'linewidth': 1})
ax3[0].text(s='2016-2019', x=ax3[0].get_xlim()[0]+6000, y=0.9, color='black',
            bbox={'boxstyle': 'Round', 'facecolor':'white', 'edgecolor':'black', 'linewidth': 1})

ax3[1].plot(euro_to_dollar_16_19['Time'], euro_to_dollar_16_19['rolling_mean'],
         label='2016 - 2019', color='grey', alpha=0.5)
ax3[1].plot(euro_to_dollar_20['Time'], euro_to_dollar_20['rolling_mean'],
         label='2020', color='black')

ax3[1].axvspan(xmin=euro_to_dollar.loc[euro_to_dollar['Time'].dt.year==2020]['Time'].min(),
            xmax=euro_to_dollar.loc[euro_to_dollar['Time'].dt.year==2020]['Time'].max(),
            alpha=1, color='#B2B2B2')
ax3[1].set(facecolor='#E5E5E5')
ax3[1].legend()

ax3[2].plot(euro_to_dollar_20['Time'], euro_to_dollar_20['rolling_mean'],
         label='2020', color='black')
ax3[2].set(facecolor='#B2B2B2')
ax3[2].legend()
```
