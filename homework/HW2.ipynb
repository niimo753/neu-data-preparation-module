# CREATING AND MANIPULATING DATAFRAMES

### Exercise 1
Create a DataFrame named life_exp_top10 with two columns Women and Men (in that order). The 
values in these columns should be the ones in respective dictionaries

```python
women = {'Hong Kong': 87.6,
         'Japan': 87.5,
         'Switzerland': 85.5,
         'Singapore': 85.7,
         'Italy': 85.4,
         'Spain': 86.1,
         'Australia': 85.3,
         'Iceland': 84.4,
         'Israel': 84.4,
         'South Korea': 85.8}
men = {'Hong Kong': 81.8,
       'Japan': 81.3,
       'Switzerland': 81.7,
       'Singapore': 81.4,
       'Italy': 81.1,
       'Spain': 80.7,
       'Australia': 81.3,
       'Iceland': 81.3,
       'Israel': 81.1,
       'South Korea': 79.7}
```

```python
life_exp_top10 = pd.DataFrame({'Women': women, 'Men': men})
life_exp_top10
```

### Exercise 2
Add a new column to life_exp_top10 named Average. Its values should be the average life expectancy of the women and men in each 

```python
life_exp_top10['Average'] = life_exp_top10.apply(lambda x: x.mean(), axis=1)
life_exp_top10
```

### Exercise 3
Add the provided rank Series as a new column of life_exp_top10 named Rank. Assign the result back to life_exp_top10

```python
rank = pd.Series([1, 2, 3, 4, 5, 6, 7, 8, 9, 9])
```

```python
rank = pd.Series(rank.values, index=life_exp_top10.index)
life_exp_top10['Rank'] = rank
life_exp_top10
```

### Exercise 4
Add a new row for Sweden. The women average life expectancy is 84.4 and the male average life 
expectancy is 80.9. The average can be calculated from these two

```python
life_exp_top10.loc['Sweden'] = {'Women': 84.4, 'Men': 80.9}
life_exp_top10.loc['Sweden', 'Average'] = life_exp_top10.loc['Sweden'].mean()
life_exp_top10
```

### Exercise 5
Remove the row containing Italy’s life expectancy data. Remember to assign the result to life_exp_top10

```python
life_exp_top10.drop('Italy', inplace=True)
life_exp_top10
```

### Question 6

```python
life_exp_top10.drop('Rank', axis=1, inplace=True)
life_exp_top10
```

### Question 7

```python
life_exp_top10 = life_exp_top10[['Average', 'Women', 'Men']]
life_exp_top10
```
