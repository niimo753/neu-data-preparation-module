### Question 2
Đổi tên cột từ định dạng camelcase sang snakecase. Ví dụ: dateCrawled &rarr; date_crawled. Gợi ý: dùng df.columns

```python
def camel_to_snake(camel):
    if camel[0].isupper():
        camel = camel[0].lower() + camel[1:]
    upper = [i for i in camel if i.isupper()]
    for i in upper:
        camel = camel.replace(i, f'_{i.lower()}')
    return camel

df.columns = [camel_to_snake(i) for i in df.columns]
df.columns
```

### Question 5
Attribute “model” chứa rất nhiều giá trị phân biệt (characteristics/unique values), tuy nhiên rất nhiều giá trị có số lượng quan sát rất ít. Gom nhóm tất cả các giá trị có số lượng quan sát chiếm tỷ lệ $\leq 0.1\%$ về 1 nhóm gọi tên là OTHER.

```python
model_freq = df['model'].value_counts()
model_freq
```

```python
total = len(df)
df['model'] = df['model'].apply(lambda x: x if x is np.nan or model_freq[x]/total>0.001 else 'other')
df.head()
```
