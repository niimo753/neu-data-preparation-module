__Function `update_vals` used to update the data__

```python
def update_vals(val):
    if val == np.nan:
        return np.nan
    elif val == '-':
        return False
    return True
```

__Apply the function__

```python
tafe_resignations[['Contributing Factors. Dissatisfaction', 'Contributing Factors. Job Dissatisfaction']] = tafe_resignations.loc[:, ('Contributing Factors. Dissatisfaction', 'Contributing Factors. Job Dissatisfaction')].applymap(update_vals)
```

__Create a `dissatisfied` column using `df.any()`__

```python
tafe_resignations['dissatisfied'] = tafe_resignations[['Contributing Factors. Dissatisfaction', 'Contributing Factors. Job Dissatisfaction']].any(axis=1)
tafe_resignations['dissatisfied'].value_counts()
```

# Combine the Data

```python
dete_resignations_up['institute'] = 'DETE'
tafe_resignations_up['institute'] = 'TAFE'
```

```python
combined = pd.concat([dete_resignations_up, tafe_resignations_up])
```

```python
# combined_updated = combined.dropna(thresh = 500, axis =1).copy()
drop_cols = [col for col in combined.columns if combined[col].notnull().sum() < 500]
combined_update = combined.drop(drop_cols, axis=1).copy()
combined_update.head()
```

__Working with the data in the `institute_service` column__

```python
combined_update['institute_service'].value_counts()
```

```python
combined_update['institute_service'] = combined_update['institute_service'].astype('str')
combined_update['institute_service'] = combined_update['institute_service'].str.extract('(\d+.\d+)')[0]
combined_update['institute_service'] = combined_update['institute_service'].str.replace('.0', '', regex=False)
combined_update['institute_service'] = combined_update['institute_service'].astype('str')
```

```python
def convert(x):
    x = x.split('-')
    if len(x) == 2:
        return (int(x[0])+int(x[1])/2)
    else:
        if x[0] == 'nan':
            return np.nan
        else:
            return float(x[0])
```

```python
combined_update['institute_service'] = combined_update['institute_service'].apply(lambda x: convert(x))
combined_update['institute_service'].value_counts()
```

__Map each year value to one of the career stages__

```python
def career_stage(year):
    if year <= 2:
        return 'Newbie'
    elif year <= 10:
        return 'Experience'
    else:
        return 'Veteran'
```

```python
combined_update['service_cat'] = combined_update['institute_service'].apply(lambda x: career_stage(x))
```
