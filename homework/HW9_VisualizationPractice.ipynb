```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### SLIDE 16

```python
df1 = pd.read_csv(r"../data/covid_avg_deaths.csv")
df1
```

```python
fig1, ax1 = plt.subplots(nrows=4, ncols=1)
period= [[0, 3, 'Jan-Mar'], [2, 6, 'Mar-Jun'], [5, 9, 'Jun-Oct'], [8, 12, 'Oct-Dec']]
for i in range(len(ax1)):
    ax1[i].plot(df1.Month, df1.New_deaths, color='r', alpha=0.2)
    
    # title and subtitle
    if i == 0:
        ax1[0].text(x=0.5, y=ax1[i].get_ylim()[1] + 1500,
                    s='The virus kills 851 people each day',
                    size=17,
                    weight='bold')
        ax1[0].text(x=0.5, y=ax1[i].get_ylim()[1] + 700,
                    s='Average number of daily deaths per month in the US',
                    size=12)
    
    # set visible
    ax1[i].spines[['left', 'right', 'top', 'bottom']].set_visible(False)
    ax1[i].tick_params(left=False, bottom=False)
    ax1[i].xaxis.set_visible(False)
    ax1[i].yaxis.set_visible(False)
    
    # highlighting
    line = ax1[i].lines[0]
    ax1[i].plot(line.get_xdata()[period[i][0]:period[i][1]],
           line.get_ydata()[period[i][0]:period[i][1]],
           color='red')
    
    # annotate
    coordinate = [(1.5, 200), (3.9, 900), (7, 1200), (9.4, 1300)]
    ax1[i].annotate(period[i][2], xy=coordinate[i])
    
    # duel axis
#     cumavg = ax1[i].twiny()
#     cumavg.spines[['left', 'right', 'top', 'bottom']].set_visible(False)
#     cumavg.tick_params(left=False, bottom=False)
#     cumavg.xaxis.set_visible(False)
#     cumavg.yaxis.set_visible(False)
#     cumavg.plot(df1.New_deaths, [2000]*len(df1))
```

```python
df2 = pd.read_excel(r"../data/Product_Price.xlsx")
df2
```

```python
df2 = df2.drop('Unnamed: 1',axis=1)
df2 = df2.rename(columns = {'Unnamed: 0':'Product Name'})
df2
```

```python
df2sub = df2.set_index('Product Name').T.copy()
fig2 ,ax2 = plt.subplots(figsize=(10,5))
ax2.plot(df2sub, linewidth=2.2, color='grey')
ax2.tick_params(color ='#DCDCDC' )
ax2.spines[['right', 'top']].set_visible(False)
ax2.spines[['left', 'bottom']].set_color('#DCDCDC')
ax2.set_title('Average Retail Product Price Per Year',size=15, color='grey')
```

```python
lines = ax2.lines

ax2.text(2007.8,390,'A',weight='bold',color ='grey',size=14)
ax2.text(2007.8,350,'B',weight='bold',color ='grey',size=14)
ax2.text(2009.8,100,'C',weight='bold',color ='grey',size=14)
ax2.text(2010.8,140,'D',weight='bold',color ='grey',size=14)
ax2.text(2012.8,98,'E',weight='bold',color ='grey',size=14)

ax2.plot(lines[0].get_xdata()[2:], lines[0].get_ydata()[2:], color='#174A7E')
ax2.plot(lines[1].get_xdata()[2:], lines[1].get_ydata()[2:], color='#174A7E')
lines[2].set_markevery([2])
lines[2].set_marker('o')
lines[2].set_markersize(9)
lines[2].set_markerfacecolor('#174A7E')

fig2
```
